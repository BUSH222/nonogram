```python
from nonogram.sieve import sieve

sizes = [5, 10, 15, 20, 25, 30, 35, 40]
frequencies = [6, 6, 6, 6, 5, 5, 5, 5]
seed_cnt = [100000, 100000, 100000, 50000, 10000, 10000, 10000, 10000]
```

```python
for size, freq, cnt in zip(sizes, frequencies, seed_cnt):
    print(f"Starting... Size: {size}, Frequency: {freq}, Seed count: {cnt}")
    results = sieve(0, cnt, size, 0.5, freq, number_limit=5, verbose=False, use_tqdm=True)
    with open(f"results_{size}_{freq}.txt", 'w') as f:
        for seed, result in results:
            f.write(f"{seed},{result}\n")
    print('Done!')
```
